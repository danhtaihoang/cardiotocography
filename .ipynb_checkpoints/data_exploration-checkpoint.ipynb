```python
# load data
df = pd.read_excel('../CTG_data.xls')
```

```python
# class (10 classes)
y1 = df.iloc[1:-3,43]
y1.head()
```

```python
# nsp (3 classes)
y2 = df.iloc[1:-3,45]
y2.head()
```

```python
y1 = np.array(y1).astype(int)
y2 = np.array(y2).astype(int)
```

```python
y1.shape
```

```python
y2.shape
```

```python
y1 = onehot_encoder.fit_transform(y1.reshape(-1,1))
y1 = np.argmax(y1,axis=1)

y2 = onehot_encoder.fit_transform(y2.reshape(-1,1))
y2 = np.argmax(y2,axis=1)
```
